### Import libraries

```python
import numpy as np 
import pandas as pd 
import librosa
import matplotlib.pyplot as plt
import os
import cv2
import IPython.display as ipd
import soundfile
```

### Load an audio file

```python
file_path = "audio_data/Audio_8hz/backward/0cb74144_nohash_2.wav"
wav, sr = librosa.load(file_path, sr=None)
print(wav.shape, wav.max(), wav.min())
```

```python
ipd.Audio(wav, rate=sr)
```

### Add background noise to the loaded audio file

```python
# add background noise code
bg_files = os.listdir('bn/')
chosen_bg_file = bg_files[np.random.randint(3)]
bg, sr2 = librosa.load('bn/'+chosen_bg_file, sr=None)

start_ = np.random.randint(bg.shape[0]-8000)
bg_slice = bg[start_ : start_+8000]
wav_with_bg = wav * np.random.uniform(0.8, 1.2) + bg_slice * np.random.uniform(0, 0.1)
```

```python
ipd.Audio(wav_with_bg, rate=sr)
```

### Speed tuning to an audio file

```python

# speed tuning
speed_rate = np.random.uniform(0.7,1.3)
wav_speed_tune = cv2.resize(wav, (1, int(len(wav) * speed_rate))).squeeze()
if len(wav_speed_tune) < 8000:
    pad_len = 8000 - len(wav_speed_tune)
    wav_speed_tune = np.r_[np.random.uniform(-0.001,0.001,int(pad_len/2)),
                           wav_speed_tune,
                           np.random.uniform(-0.001,0.001,int(np.ceil(pad_len/2)))]
else: 
    cut_len = len(wav_speed_tune) - 8000
    wav_speed_tune = wav_speed_tune[int(cut_len/2):int(cut_len/2)+8000]
```

```python
ipd.Audio(wav_speed_tune, rate=sr)
```

### Adding background noise to 500 samples of each class

```python
train_audio_path = "audio_data/Audio_8hz"
labels=os.listdir(train_audio_path)
```

```python
## To add background noise

labels = ['nine']
for label in labels:
    count = 0
    dir_path = "audio_data/Audio_8hz/" + label
    print(label)
    for file in os.listdir(dir_path):
        file_path = dir_path+"/"+file
        wav, sr1 = librosa.load(file_path, sr=None)
        if len(wav) == 8000:
            print(file_path)
            # add background noise code
            bg_files = os.listdir('bn/')
            chosen_bg_file = bg_files[np.random.randint(3)]
            bg, sr2 = librosa.load('bn/'+chosen_bg_file, sr=None)

            start_ = np.random.randint(bg.shape[0]-8000)
            bg_slice = bg[start_ : start_+8000]
            wav_with_bg = wav * np.random.uniform(0.8, 1.2) + bg_slice * np.random.uniform(0, 0.1)
            soundfile.write(file_path, wav_with_bg, sr1)
            count = count + 1
            if count >= 500:
                break
```

### Speed tuning of 200 samples of audio files of each class

```python
## Speed tuning

for label in labels:
    count = 0
    dir_path = "audio_data/Audio_8hz/" + label
    print(label)
    for file in os.listdir(dir_path):
        count = count + 1
        if(count < 500):
            continue
        file_path = dir_path + "/" + file
        wav, sr1 = librosa.load(file_path, sr=None)
        if len(wav) == 8000:
            print(file_path)
            speed_rate = np.random.uniform(0.7,1.3)
            wav_speed_tune = cv2.resize(wav, (1, int(len(wav) * speed_rate))).squeeze()
            if len(wav_speed_tune) < 8000:
                pad_len = 8000 - len(wav_speed_tune)
                wav_speed_tune = np.r_[np.random.uniform(-0.001,0.001,int(pad_len/2)),
                                       wav_speed_tune,
                                       np.random.uniform(-0.001,0.001,int(np.ceil(pad_len/2)))]
            else: 
                cut_len = len(wav_speed_tune) - 8000
                wav_speed_tune = wav_speed_tune[int(cut_len/2):int(cut_len/2)+8000]
            soundfile.write(file_path, wav_speed_tune, sr1)

            if count >= 700:
                break
```
